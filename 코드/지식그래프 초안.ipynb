```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 윈도우의 경우 한글 폰트 경로 설정
font_path = 'C:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 그래프 초기화
G = nx.Graph()

# 노드 추가 (경력, 스킬, 프로젝트, 산업군)
jobs = ["한미약품 - 데이터 분석가", "따릉이 데이터 분석가", "부동산 가격 예측", "보이스피싱 분석"]
skills = ["파이썬", "SAS", "엑셀", "QGIS", "머신러닝", "통계"]
projects = ["따릉이 데이터", "부동산 데이터", "보이스피싱 데이터"]
industries = ["제약", "금융", "공공 서비스", "부동산"]

# 경력 노드 추가
for job in jobs:
    G.add_node(job, label="Job")

# 스킬 노드 추가
for skill in skills:
    G.add_node(skill, label="Skill")

# 프로젝트 노드 추가
for project in projects:
    G.add_node(project, label="Project")

# 산업군 노드 추가
for industry in industries:
    G.add_node(industry, label="Industry")

# 경력-스킬 간의 연결 (엣지 추가)
G.add_edges_from([
    ("한미약품 - 데이터 분석가", "SAS"),
    ("한미약품 - 데이터 분석가", "엑셀"),
    ("따릉이 데이터 분석가", "파이썬"),
    ("부동산 가격 예측", "파이썬"),
    ("부동산 가격 예측", "QGIS"),
    ("보이스피싱 분석", "파이썬"),
    ("보이스피싱 분석", "머신러닝")
])

# 경력-프로젝트 간의 연결 (엣지 추가)
G.add_edges_from([
    ("따릉이 데이터 분석가", "따릉이 데이터"),
    ("부동산 가격 예측", "부동산 데이터"),
    ("보이스피싱 분석", "보이스피싱 데이터")
])

# 경력-산업군 간의 연결 (엣지 추가)
G.add_edges_from([
    ("한미약품 - 데이터 분석가", "제약"),
    ("따릉이 데이터 분석가", "공공 서비스"),
    ("부동산 가격 예측", "부동산"),
    ("보이스피싱 분석", "금융")
])

# 그래프 시각화
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color="lightgreen", font_size=10, node_size=3000, font_weight="bold")
plt.title("정호원님의 경력 관련 지식 그래프", fontsize=15)
plt.show()
```
