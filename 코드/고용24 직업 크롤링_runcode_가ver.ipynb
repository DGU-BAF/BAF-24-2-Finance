```python
# 페이지 목록이 11 이상인 경우 사용
# 페이지 목록이 10 이하인 경우에는 코드_나.ver를 사용해주세요

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import csv

# ChromeDriver 경로 설정
chrome_driver_path = "C:/Users/chica/OneDrive/바탕 화면/chromedriver-win64/chromedriver.exe"

# WebDriver 옵션 설정
options = Options()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36")

# WebDriver 초기화
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=options)
driver.implicitly_wait(3)

# 페이지 불러오기
driver.get('https://www.work.go.kr/consltJobCarpa/srch/jobDic/getJobDicSrchByInitWord.do?pageType=jobDicIntro&txtNumber=6&initWord=&pageIndex=1&pageUnit=10&pageSize=10&stardWord=%EB%B0%94&endWord=%EC%82%AC')  # 직업 사이트

try:

    # 수집한 데이터를 CSV로 작성할 때, UTF-8 인코딩 확인
    with open("../데이터/직업정보크롤링_바.csv", 'w', newline='', encoding='utf-8-sig') as f:
        writer = csv.writer(f)
        writer.writerow(['직무명', '직무개요', '수행직무', '정규교육', '숙련기간', '직무기능', '작업강도', '육체활동', '작업장소', '작업환경', '유사명칭', '관련직업', '자격/면허', '고용직업분류', '표준직업분류', '표준산업분류', '조사연도', '비고'])  # CSV 헤더 설정

        page_num = 1  # 현재 페이지 번호

        while page_num <= 47:  # 47페이지까지 크롤링 (필요한 페이지 수로 수정)
            item_list_xpath = '//*[@id="contents"]/div[5]/div/table/tbody/tr/td/ul/li'
            items = driver.find_elements(By.XPATH, item_list_xpath)
            num_items = len(items)  # 해당 페이지의 항목 수

            for i in range(1, num_items + 1):  # 수집한 항목 수만큼 반복
                try:
                    # XPath를 이용해 각 항목 클릭
                    element_xpath = f'//*[@id="contents"]/div[5]/div/table/tbody/tr/td/ul/li[{i}]/a'
                    element = driver.find_element(By.XPATH, element_xpath)
                    element.click()
                    time.sleep(2)  # 페이지 로딩 대기 시간

                    # 필요한 정보 수집
                    title_xpath = '//*[@id="contents"]/div[4]/div[2]'  # 직무명
                    desc1_xpath = '//*[@id="contents"]/div[5]/div[1]'  # 직무개요
                    desc2_xpath = '//*[@id="contents"]/div[5]/div[2]'  # 수행직무

                    # 각 부가직업정보 XPath
                    education_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[1]'
                    training_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[2]'
                    job_function_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[3]'
                    work_intensity_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[4]'
                    physical_activity_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[5]'
                    work_place_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[6]'
                    work_environment_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[7]'
                    similar_name_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[8]'
                    related_job_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[9]'
                    qualifications_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[10]'
                    employment_classification_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[11]'
                    standard_occupational_classification_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[12]'
                    standard_industrial_classification_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[13]'
                    survey_year_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[14]'
                    note_xpath = '//*[@id="contents"]/div[5]/div[3]/ul/li[15]'

                    # 직무명
                    title_element = driver.find_element(By.XPATH, title_xpath)
                    title = title_element.text if title_element else ''

                    # 직무개요
                    desc1_element = driver.find_element(By.XPATH, desc1_xpath)
                    description1 = desc1_element.text if desc1_element else ''

                    # 수행직무
                    desc2_element = driver.find_element(By.XPATH, desc2_xpath)
                    description2 = desc2_element.text if desc2_element else ''

                    # ':' 뒤의 텍스트만 파싱
                    def get_text_after_colon(xpath):
                        element = driver.find_element(By.XPATH, xpath)
                        full_text = element.text if element else ''
                        return full_text.split(":")[1].strip() if ':' in full_text else ''

                    # 파싱된 텍스트만 가져옴
                    education = get_text_after_colon(education_xpath)
                    training = get_text_after_colon(training_xpath)
                    job_function = get_text_after_colon(job_function_xpath)
                    work_intensity = get_text_after_colon(work_intensity_xpath)
                    physical_activity = get_text_after_colon(physical_activity_xpath)
                    work_place = get_text_after_colon(work_place_xpath)
                    work_environment = get_text_after_colon(work_environment_xpath)
                    similar_name = get_text_after_colon(similar_name_xpath)
                    related_job = get_text_after_colon(related_job_xpath)
                    qualifications = get_text_after_colon(qualifications_xpath)
                    employment_classification = get_text_after_colon(employment_classification_xpath)
                    standard_occupational_classification = get_text_after_colon(standard_occupational_classification_xpath)
                    standard_industrial_classification = get_text_after_colon(standard_industrial_classification_xpath)
                    survey_year = get_text_after_colon(survey_year_xpath)
                    note = get_text_after_colon(note_xpath)

                    # 수집한 데이터 CSV에 작성
                    writer.writerow([title, description1, description2, education, training, job_function, work_intensity, physical_activity, work_place, work_environment, similar_name, related_job, qualifications, employment_classification, standard_occupational_classification, standard_industrial_classification, survey_year, note])

                    # 뒤로 가기
                    driver.back()
                    time.sleep(2)
                except Exception as e:
                    print(f"Error processing item {i} on page {page_num}: {e}")

            # 페이지 그룹 변경 및 다음 페이지로 이동하는 부분
            try:
                if page_num % 10 == 0:  # 페이지가 10묶음씩 있음
                    next_button_xpath = '//*[@id="contents"]/nav/a[12]'  # 모든 페이지에서 다음버튼은 [12]로 고정됨
                else:
                    next_button_xpath = f'//*[@id="contents"]/nav/a[{(page_num % 10) + 2}]'  # 다음 페이지로 이동하는 버튼(변수가 1이후로는 3부터 나열됨. 따라서 2증가)
                next_button = driver.find_element(By.XPATH, next_button_xpath)
                next_button.click()
                time.sleep(5)  # 페이지 로딩 대기 시간
            except Exception as e:
                print(f"Error navigating to next page: {e}")
                break  # 페이지 이동 중 오류 발생 시 종료

            # 페이지 번호 증가
            page_num += 1

finally:
    driver.quit()
```
