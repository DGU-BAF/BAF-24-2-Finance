### 직업코드 수정1

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import csv

# ChromeDriver 경로 설정
chrome_driver_path = 'C:/Users/정호원/OneDrive/바탕 화면/chromedriver-win64/chromedriver.exe'

# WebDriver 옵션 설정
options = Options()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36")

# WebDriver 초기화
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=options)
driver.implicitly_wait(3)

# 크롤링할 페이지 URL
url = 'https://www.work.go.kr/consltJobCarpa/srch/jobDic/getJobDicSrchByInitWord.do?pageType=jobDicIntro&txtNumber=1&initWord=&pageIndex=1&pageUnit=10&pageSize=10&stardWord=%EA%B0%80&endWord=%EB%82%98'
driver.get(url)

try:
    # CSV 파일 열기
    with open('C:/Users/정호원/OneDrive/바탕 화면/직업크롤링.csv', 'w', newline='', encoding='UTF-8') as f:
        writer = csv.writer(f)
        writer.writerow(['Title', 'Description 1', 'Description 2', 'Tags'])  # CSV 헤더 설정

        # 페이지 인덱스 및 페이지 넘김 관련 변수 설정
        page_group = 1  # 페이지 그룹 인덱스 (예: 1~10, 11~20 등)
        current_page_index = 3  # 현재 그룹 내에서의 페이지 인덱스 (3부터 시작)

        while page_group <= 2:  # 페이지 그룹을 2개까지만 크롤링 (원하는 만큼 조정 가능)
            while current_page_index <= 11:  # 페이지 그룹 내에서 3번~11번까지 반복
                for i in range(1, 21):  # 한 페이지당 20개의 항목 클릭
                    try:
                        # XPath를 이용해 각 항목 클릭
                        element_xpath = f'//*[@id="contents"]/div[5]/div/table/tbody/tr/td/ul/li[{i}]/a'
                        element = driver.find_element(By.XPATH, element_xpath)
                        element.click()
                        time.sleep(2)  # 페이지 로딩 대기 시간

                        # 필요한 정보 수집
                        title_xpath = '//*[@id="contents"]/div[5]/div[1]'  # 제목 수집
                        desc1_xpath = '//*[@id="contents"]/div[5]/div[2]'  # 설명1 수집
                        desc2_xpath = '//*[@id="contents"]/div[4]/div[2]'  # 추가 설명 수집
                        tags_xpath_base = '//*[@id="contents"]/div[5]/div[3]/ul/li'  # 태그 수집 base path

                        # 타이틀
                        title_element = driver.find_element(By.XPATH, title_xpath)
                        title = title_element.text if title_element else ''

                        # 설명1
                        desc1_element = driver.find_element(By.XPATH, desc1_xpath)
                        description1 = desc1_element.text if desc1_element else ''

                        # 설명2 (추가 설명)
                        desc2_element = driver.find_element(By.XPATH, desc2_xpath)
                        description2 = desc2_element.text if desc2_element else ''

                        # 태그 수집 (1~15개의 태그)
                        tags_list = []
                        for tag_index in range(1, 16):  # li[1] ~ li[15] 까지 수집
                            tag_xpath = f'{tags_xpath_base}[{tag_index}]'
                            try:
                                tag_element = driver.find_element(By.XPATH, tag_xpath)
                                if tag_element:
                                    tags_list.append(tag_element.text)
                            except Exception as e:
                                print(f"Error collecting tag {tag_index}: {e}")
                        
                        tags = ' '.join(tags_list) if tags_list else ''

                        # 수집한 데이터 CSV에 작성
                        writer.writerow([title, description1, description2, tags])

                        # 뒤로 가기
                        driver.back()
                        time.sleep(2)
                    except Exception as e:
                        print(f"Error processing item {i} on page {current_page_index - 2}: {e}")

                # 페이지 이동
                try:
                    # 다음 페이지 버튼 클릭
                    page_button_xpath = f'//*[@id="contents"]/nav/a[{current_page_index}]'
                    page_button = driver.find_element(By.XPATH, page_button_xpath)
                    page_button.click()
                    time.sleep(5)  # 페이지 로딩 대기 시간
                except Exception as e:
                    print(f"Error navigating to page {current_page_index - 2} in group {page_group}: {e}")

                # 다음 페이지로 이동하기 위해 인덱스 증가
                current_page_index += 1

            # '다음' 버튼 클릭하여 다음 페이지 그룹으로 이동
            try:
                next_button_xpath = '//*[@id="contents"]/nav/a[12]'  # '다음' 버튼의 XPath
                next_button = driver.find_element(By.XPATH, next_button_xpath)
                next_button.click()
                time.sleep(5)  # 페이지 로딩 대기 시간
            except Exception as e:
                print(f"Error clicking 'Next' button: {e}")

            # 페이지 그룹을 다음 그룹으로 이동하고, 인덱스 초기화
            page_group += 1
            current_page_index = 3  # 다시 페이지 인덱스를 3부터 시작

finally:
    driver.quit()
```
